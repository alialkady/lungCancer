```python
from sklearn import preprocessing

# Convert categorical --> numbers
LE=preprocessing.LabelEncoder()

df['patient_id']=LE.fit_transform(df['patient_id'])
df['gender']=LE.fit_transform(df['gender'])
df['treatment_type']=LE.fit_transform(df['treatment_type'])
df['level'] = df['level'].map({'High': 2, 'Medium': 1, 'Low': 0})
df
```

# Feature Selection

```python
from sklearn import preprocessing
from sklearn.feature_selection import SelectKBest, chi2
```

```python
feature_columns = df.columns.drop(['patient_id', 'level', 'treatment_type']) 
X = df[feature_columns]

y= df['level']
```

```python
# Cell 3: Feature selection
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd

# Initialize SelectKBest with chi2
k = 10  # Number of top features to select
selector = SelectKBest(score_func=chi2, k=k)

# Fit the selector to the data
selector.fit(X, y)

# Get the scores for each feature
scores = selector.scores_

# Create a DataFrame to display feature scores
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': scores})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

# Select top k features based on the score
top_features = feature_scores['Feature'].head(k).tolist()

print("Top 10 Features Selected:\n", top_features)

# Now modify your DataFrame to include only these features and the 'level' column
df = df[top_features + ['level']]

# Display the first few rows of the selected DataFrame to verify
print(df.head())
```

```python
# Select top k features
top_features = feature_scores.head(k)

# Plot the scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Feature', data=top_features, palette='viridis')
plt.title(f'Top {k} Features by Chi-Squared Score')
plt.xlabel('Chi-Squared Score')
plt.ylabel('Feature')
plt.show()
```

# Logistic Regression Model


```python
# Get the columns selected by SelectKBest
selected_features = feature_scores['Feature'].head(k).tolist()

# Create a new DataFrame with selected features
X_selected = X[selected_features]

print(f"Selected Features ({k}): {selected_features}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Initialize the model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)


# Evaluate the model
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Random Forest Model

```python
# Import Necessary Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
```

```python
# 2. Initialize the Random Forest Model
rf_model = RandomForestClassifier(n_estimators=20, random_state=42)

# 3. Train the Model
rf_model.fit(X_train, y_train)

print("Random Forest model has been trained on the training data.\n")

# 4. Make Predictions
y_rf_pred = rf_model.predict(X_test)

print("Predictions made on the test data.\n")

# 5. Evaluate the Model
# Checking how good the model's guesses are by comparing them to the actual values.
conf_matrix = confusion_matrix(y_test, y_rf_pred)
class_report = classification_report(y_test, y_rf_pred)

print("Random Forest Confusion Matrix:\n", conf_matrix)
print("\nRandom Forest Classification Report:\n", class_report)
```

```python

# 6. Diagnose and Address the 100% Accuracy Issue

# A. Verify Data Splitting
# Ensure that training and testing data are properly separated.
print("Verifying data splitting...")
print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}\n")

# B. Check for Data Leakage
# Make sure no information from the test set is used in training.
print("Checking for data leakage by listing feature columns:")
print(X_selected.columns.tolist(), "\n")



# C. Implement Cross-Validation
# Using cross-validation to see how the model performs on different parts of the data.
print("Performing 5-Fold Cross-Validation...")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(rf_model, X_selected, y, cv=cv, scoring='accuracy')

print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean(), "\n")



# D. Evaluate on Training vs. Testing Data
# Compare how well the model performs on both training and testing sets.
print("Evaluating model performance on training and testing data...")
y_train_pred = rf_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_rf_pred)

print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Testing Accuracy: {test_accuracy:.2f}\n")



# E. Examine Feature Importance
# See which features are most important in making predictions.
print("Analyzing feature importance...")
importances = rf_model.feature_importances_
feature_importance_rf = pd.DataFrame({
    'Feature': X_selected.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feature_importance_rf, "\n")

# Plotting Feature Importance
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_rf, palette='viridis')
plt.title('Feature Importance from Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()



# F. Check for Class Imbalance
# Ensure that the classes are balanced to prevent the model from being biased.
print("Checking for class imbalance in the target variable:")
print(y.value_counts(), "\n")


# If classes are imbalanced, consider techniques like resampling or using class weights.

# G. Ensure No Identifiers Are Included as Features
# Make sure that unique identifiers like 'patient_id' are not used as features.
print("Ensuring no identifiers are included as features:")
print("Features:", X_selected.columns.tolist(), "\n")


# Confirm that 'patient_id' is not in the list.

# H. Calculate ROC-AUC Score (Optional)
# Measure the ability of the model to distinguish between classes.
print("Calculating ROC-AUC Score...")
roc_auc = roc_auc_score(y_test, rf_model.predict_proba(X_test), multi_class='ovr')
print("ROC-AUC Score:", roc_auc, "\n")


# I. Re-train with Hyperparameter Tuning (Optional)
# Fine-tune the model to see if performance can be improved.
print("Starting Hyperparameter Tuning with GridSearchCV...")
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=1,
    scoring='accuracy'
)

# Fit GridSearchCV to the data
grid_search.fit(X_train, y_train)

# Display the best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", grid_search.best_score_, "\n")

# Evaluate the best model
best_rf_model = grid_search.best_estimator_
y_best_rf_pred = best_rf_model.predict(X_test)

print("Best Random Forest Confusion Matrix:\n", confusion_matrix(y_test, y_best_rf_pred))
print("\nBest Random Forest Classification Report:\n", classification_report(y_test, y_best_rf_pred))



```

# Support Vector Machine (SVM) Model

```python
# Import libraries for SVM and evaluation
from sklearn.svm import SVC
```

```python
# Initialize the SVM model
svm_model = SVC(kernel='linear', probability=True, random_state=42)
```

```python
# Train the SVM model
svm_model.fit(X_train, y_train)

# Make predictions with the SVM model
y_svm_pred = svm_model.predict(X_test)

# Evaluate the SVM model
conf_matrix = confusion_matrix(y_test, y_svm_pred)
class_report = classification_report(y_test, y_svm_pred)

print("SVM Confusion Matrix:\n", conf_matrix)
print("\nSVM Classification Report:\n", class_report)
```

# XGBoost Model

```python
import xgboost as xgb
from xgboost import XGBClassifier
```

```python


# Initialize the XGBoost Classifier
xgb_model = XGBClassifier(
    objective='multi:softprob',
    num_class=3,
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42
)

# Train the model
xgb_model.fit(X_train, y_train)

print("XGBoost model has been trained on the training data.\n")

# Make predictions on the test set
y_xgb_pred = xgb_model.predict(X_test)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_xgb_pred)
print("XGBoost Confusion Matrix:\n", conf_matrix, "\n")

# Classification Report
class_report = classification_report(y_test, y_xgb_pred)
print("XGBoost Classification Report:\n", class_report)

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_xgb_pred)
print(f"XGBoost Accuracy: {accuracy:.2f}\n")

# Calculate ROC-AUC Score
y_xgb_prob = xgb_model.predict_proba(X_test)
roc_auc = roc_auc_score(y_test, y_xgb_prob, multi_class='ovr')
print(f"XGBoost ROC-AUC Score: {roc_auc:.2f}\n")

# Feature Importances
importances = xgb_model.feature_importances_
feature_importance_xgb = pd.DataFrame({
    'Feature': X_selected.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("XGBoost Feature Importances:\n", feature_importance_xgb, "\n")

# Plot Feature Importances
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_xgb, palette='viridis')
plt.title('Feature Importance from XGBoost')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# Optional: Hyperparameter Tuning with GridSearchCV
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

grid_search = GridSearchCV(
    estimator=XGBClassifier(
        objective='multi:softprob',
        num_class=3,
        eval_metric='mlogloss',
        use_label_encoder=False,
        random_state=42
    ),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("Best Parameters from GridSearchCV:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_, "\n")

# Evaluate the best model
best_xgb_model = grid_search.best_estimator_
y_best_xgb_pred = best_xgb_model.predict(X_test)

print("Best XGBoost Confusion Matrix:\n", confusion_matrix(y_test, y_best_xgb_pred))
print("\nBest XGBoost Classification Report:\n", classification_report(y_test, y_best_xgb_pred))
```
